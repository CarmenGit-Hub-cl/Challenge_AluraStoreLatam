#1. Análisis de facturación



Antes de pasar a análisis detallados, es esencial explorar el conjunto de datos para comprender su estructura y contenido. Este paso le permite identificar patrones, inconsistencias y las columnas más relevantes para los siguientes pasos.

Estructura de datos:
El conjunto de datos incluye la siguiente información:

Producto y Categoría: Artículos vendidos y sus calificaciones.

Precio y Envío: Valores de venta y costos asociados.

Fecha y ubicación de compra: Información temporal y geográfica.

Evaluación de compra: Comentarios de clientes.

Tipo de Pago y Cuotas: Métodos utilizados por los clientes.

Coordenadas Geográficas: Ubicación de las transacciones.



# **Ingreso total por cada tienda**

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingreso_tienda = tienda["Precio"].sum()
print(f"Ingreso total tienda : {ingreso_tienda:,.0f}".replace(",", "."))
ingreso_tienda2 = tienda2["Precio"].sum()
print(f"Ingreso total tienda2: {ingreso_tienda2:,.0f}".replace(",", "."))
ingreso_tienda3 = tienda3["Precio"].sum()
print(f"Ingreso total tienda3: {ingreso_tienda3:,.0f}".replace(",", "."))
ingreso_tienda4 = tienda4["Precio"].sum()
print(f"Ingreso total tienda4: {ingreso_tienda4:,.0f}".replace(",", "."))
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
categoria_popular_tienda = tienda.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)
categoria_popular_tienda
```

```python
categoria_popular_tienda2 = tienda2.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)
categoria_popular_tienda2
```

```python
categoria_popular_tienda3 = tienda3.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)
categoria_popular_tienda3
```

```python
categoria_popular_tienda4 = tienda4.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)
categoria_popular_tienda4
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
print(f'Calificación promedio tienda : {tienda.Calificación.mean():,.2f}')
print(f'Calificación promedio tienda2: {tienda2.Calificación.mean():,.2f}')
print(f'Calificación promedio tienda3: {tienda3.Calificación.mean():,.2f}')
print(f'Calificación promedio tienda4: {tienda4.Calificación.mean():,.2f}')
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué **productos** fueron los más vendidos y los menos vendidos en cada tienda. Para ello, se debe agrupar por la columna **Producto** y contar cuántas veces aparece cada uno. Esto permite detectar cuáles artículos tienen mayor rotación y cuáles presentan menor demanda.

A continuación se calculan los productos más y menos vendidos de cada tienda:

```python
# Tienda 1
ventas_producto_t1 = tienda["Producto"].value_counts()
print("Tienda 1")
print(f"Producto más vendido: {ventas_producto_t1.idxmax()} ({ventas_producto_t1.max()} ventas)")
print(f"Producto menos vendido: {ventas_producto_t1.idxmin()} ({ventas_producto_t1.min()} ventas)")
```

```python
# Tienda 2
ventas_producto_t2 = tienda2["Producto"].value_counts()
print("Tienda 2")
print(f"Producto más vendido: {ventas_producto_t2.idxmax()} ({ventas_producto_t2.max()} ventas)")
print(f"Producto menos vendido: {ventas_producto_t2.idxmin()} ({ventas_producto_t2.min()} ventas)")
```

```python
# Tienda 3
ventas_producto_t3 = tienda3["Producto"].value_counts()
print("Tienda 3")
print(f"Producto más vendido: {ventas_producto_t3.idxmax()} ({ventas_producto_t3.max()} ventas)")
print(f"Producto menos vendido: {ventas_producto_t3.idxmin()} ({ventas_producto_t3.min()} ventas)")
```

```python
# Tienda 4
ventas_producto_t4 = tienda4["Producto"].value_counts()
print("Tienda 4")
print(f"Producto más vendido: {ventas_producto_t4.idxmax()} ({ventas_producto_t4.max()} ventas)")
print(f"Producto menos vendido: {ventas_producto_t4.idxmin()} ({ventas_producto_t4.min()} ventas)")
```

# 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python

prm_envio_tienda = tienda['Costo de envío'].mean()
print(f"Promedio envío tienda : {prm_envio_tienda:,.2f}")
prm_envio_tienda2 = tienda2['Costo de envío'].mean()
print(f"Promedio envío tienda2 : {prm_envio_tienda2:,.2f}")
prm_envio_tienda3 = tienda3['Costo de envío'].mean()
print(f"Promedio envío tienda3 : {prm_envio_tienda3:,.2f}")
prm_envio_tienda4 = tienda4['Costo de envío'].mean()
print(f"Promedio envío tienda4 : {prm_envio_tienda4:,.2f}")
```

# **Generando gráfico**

Después de realizar los análisis, es hora de transformar sus resultados en visualizaciones que le ayuden a comprender mejor los patrones y los insights encontrados.

De acuerdo con funciones que haya creado, le recomendamos que genere al menos tres gráficos. Estos gráficos deben ser de diferentes tipos (como barras, líneas, dispersión, entre otros) para presentar una visión completa de los datos.

Recuerde que los gráficos deben complementar el análisis realizado, resaltando los puntos más relevantes, como los ingresos de la tienda, la distribución de categorías de productos, las opiniones de los clientes, los productos más vendidos y/o los costes de envío.

No es necesario mostrar una visualización en cada paso, pero al final del análisis debes generar los gráficos que consideres más útiles para interpretar los datos.

Consejo: Matplotlib ofrece una variedad de tipos de gráficos que son fáciles de implementar, lo que lo convierte en una excelente opción para principiantes. Utilice su creatividad para elegir los gráficos que mejor resalten las tendencias y los conocimientos de su análisis.

```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
promedios = [prm_envio_tienda, prm_envio_tienda2, prm_envio_tienda3, prm_envio_tienda4]

# Crear gráfico de barras
plt.figure(figsize=(8,4))
plt.bar(tiendas, promedios, color=['#4e79a7','#f28e2b','#e15759','#76b7b2'])
plt.title('Promedio de Costo de Envío por Tienda', fontsize=14)
plt.xlabel('Tiendas')
plt.ylabel('Costo promedio de envío ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar valores sobre cada barra
for i, v in enumerate(promedios):
    plt.text(i, v + 0.05*v, f"${v:,.2f}", ha='center', fontsize=10)

plt.show()
```

```python

# Crear listas para el gráfico
etiquetas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
valores = [ingreso_tienda, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

# Gráfico de pastel
plt.figure(figsize=(6, 6))
plt.pie(
    valores,
    labels=etiquetas,
    autopct='%1.1f%%',    # Muestra los porcentajes con un decimal
    startangle=90,        # Comienza el gráfico desde arriba
    colors=['#66b3ff', '#99ff99', '#ffcc99', '#ff9999'],  # Colores personalizados
    shadow=True
)

plt.title('Participación del Ingreso Total por Tienda', fontsize=14)
plt.show()
```

```python
# Unir los resultados en un solo DataFrame

productos_tienda = tienda.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)#).to_frame() #para convetir a dataframe

productos_tienda2 = tienda2.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)#).to_frame() #para convetir a dataframe

productos_tienda3 = tienda3.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)#).to_frame() #para convetir a dataframe

productos_tienda4 = tienda4.groupby("Categoría del Producto")["Categoría del Producto"].count().sort_values(ascending=False)#).to_frame() #para convetir a dataframe


comparacion_productos = pd.concat(
    [productos_tienda, productos_tienda2, productos_tienda3, productos_tienda4],
    axis=1
)
comparacion_productos.columns = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
comparacion_productos = comparacion_productos.fillna(0)  # Reemplazar NaN por 0

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
comparacion_productos.plot(kind='bar', figsize=(12, 6))

plt.title('Comparación de Cantidad de Productos Vendidos por Categoría y Tienda', fontsize=14)
plt.xlabel('Categoría del Producto', fontsize=12)
plt.ylabel('Cantidad de Productos', fontsize=12)
plt.legend(title='Tienda')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Informe final para el señor Juan

Con base en el análisis de las cuatro tiendas, se evaluaron los siguientes criterios:

1. **Facturación total**
2. **Categorías más populares**
3. **Calificación promedio de los clientes**
4. **Productos más y menos vendidos**
5. **Costo promedio de envío**

El objetivo es identificar cuál tienda presenta el desempeño más bajo para recomendar cuál debería venderse e invertir ese capital en un nuevo negocio.


```python
# Resumen comparativo final
resumen = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Facturación total": [ingreso_tienda, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4],
    "Calificación promedio": [tienda.Calificación.mean(), tienda2.Calificación.mean(), tienda3.Calificación.mean(), tienda4.Calificación.mean()],
    "Costo promedio de envío": [prm_envio_tienda, prm_envio_tienda2, prm_envio_tienda3, prm_envio_tienda4],
    "Categoría más popular": [
        categoria_popular_tienda.idxmax(),
        categoria_popular_tienda2.idxmax(),
        categoria_popular_tienda3.idxmax(),
        categoria_popular_tienda4.idxmax()
    ],
    "Producto más vendido": [
        ventas_producto_t1.idxmax(),
        ventas_producto_t2.idxmax(),
        ventas_producto_t3.idxmax(),
        ventas_producto_t4.idxmax()
    ],
    "Producto menos vendido": [
        ventas_producto_t1.idxmin(),
        ventas_producto_t2.idxmin(),
        ventas_producto_t3.idxmin(),
        ventas_producto_t4.idxmin()
    ]
})

resumen
```

## Conclusión y recomendación

Después de comparar los resultados, la **Tienda 4** es la que muestra el desempeño más bajo en términos generales, por las siguientes razones:

- Tiene la **menor facturación total** entre las cuatro tiendas.
- Aunque su **calificación promedio** no es la peor, tampoco compensa su menor nivel de ingresos.
- Presenta el **menor costo promedio de envío**, lo cual es positivo, pero este aspecto por sí solo no logra equilibrar su menor facturación.
- En categorías populares mantiene un comportamiento similar al de las demás tiendas, pero no destaca en ventas totales frente a sus competidoras.

### Recomendación al señor Juan

Se recomienda al **señor Juan considerar la venta de la Tienda 4**, ya que es la que genera **menor ingreso total** y, en una evaluación conjunta de los indicadores analizados, presenta el rendimiento más bajo para el objetivo estratégico planteado.

Esta decisión permitiría liberar recursos para invertir en un nuevo negocio con mayor potencial de rentabilidad.
